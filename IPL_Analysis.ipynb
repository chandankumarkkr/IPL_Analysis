### Find the most destructive death-over(last 5 overs) batsman(who has played at least 200 balls) in the history of IPL
### Run rate = (total_run/total no.of ball played)*100

```python
death_over=deliveries[deliveries['over']>15]
```

```python
All_in_death_over=death_over.groupby('batsman')['batsman_runs'].count()
x=All_in_death_over>200
```

```python
## Masking will work only on same size elements
batsman_list=All_in_death_over[x].index.tolist()
```

```python
#runs scored by all the batsman in list
# balls played by all the batsman in the list

final=death_over[death_over['batsman'].isin(batsman_list)]
```

```python
runs=final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls=final.groupby('batsman')['batsman_runs'].count()
```

```python
strike_rate=(runs/balls)*100
```

```python
strike_rate
```
